```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by Category and sum the Sales
category_sales = data.groupby('Category')['SalesperCustomer'].sum().reset_index()

# Create bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='SalesperCustomer', data=category_sales)
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Group data by Country and sum the Sales
country_sales = data.groupby('Country')['SalesperCustomer'].sum()

# Create pie chart
plt.figure(figsize=(10, 6))
country_sales.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Country')
plt.ylabel('')
plt.show()
```

```python
# Create box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Sub_Category', y='Profit', data=data)
plt.title('Profit by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by OrderDate and sum the Sales
time_series = data.groupby('OrderDate')['SalesperCustomer'].sum().reset_index()

# Create line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='OrderDate', y='SalesperCustomer', data=time_series)
plt.title('Sales Trends over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load Data from Different File Formats
sales_csv = pd.read_csv("sales-data-sample.csv")
sales_excel = pd.read_excel("sales-data-sample.xlsx")
sales_json = pd.read_json("sales-data-sample.json")

# Step 2: Explore Data Structure and Content
print("CSV Columns:", sales_csv.columns.tolist())
print("Excel Columns:", sales_excel.columns.tolist())
print("JSON Columns:", sales_json.columns.tolist())

print("CSV Data Sample:\n", sales_csv.head())
print("Excel Data Sample:\n", sales_excel.head())
print("JSON Data Sample:\n", sales_json.head())

# Step 3: Data Cleaning
sales_csv = sales_csv.fillna(0)
sales_excel = sales_excel.fillna(0)
sales_json = sales_json.fillna(0)

sales_csv.drop_duplicates(inplace=True)
sales_excel.drop_duplicates(inplace=True)
sales_json.drop_duplicates(inplace=True)

# Step 4: Standardize column names (lowercase + remove spaces)
def clean_columns(df):
    df.columns = df.columns.str.lower().str.replace(" ", "").str.replace("_", "")
    return df

sales_csv = clean_columns(sales_csv)
sales_excel = clean_columns(sales_excel)
sales_json = clean_columns(sales_json)

# Step 4b: Rename inconsistent columns to a common format
def rename_columns(df):
    mapping = {}
    for col in df.columns:
        if "quantity" in col:
            mapping[col] = "quantity"
        elif "price" in col:
            mapping[col] = "price"
        elif "product" in col or "category" in col or "line" in col:
            mapping[col] = "product_category"
    df.rename(columns=mapping, inplace=True)
    return df

sales_csv = rename_columns(sales_csv)
sales_excel = rename_columns(sales_excel)
sales_json = rename_columns(sales_json)

# Step 5: Combine into a single dataframe
all_sales_data = pd.concat([sales_csv, sales_excel, sales_json], ignore_index=True)
print("Unified Columns:", all_sales_data.columns.tolist())

# Step 6: Data Transformation
if "quantity" in all_sales_data.columns and "price" in all_sales_data.columns:
    all_sales_data["order_value"] = pd.to_numeric(all_sales_data["quantity"], errors="coerce") * pd.to_numeric(all_sales_data["price"], errors="coerce")
else:
    raise ValueError("Required columns 'quantity' and 'price' are missing!")

# Step 7: Analyze Sales Data
print(all_sales_data.describe())

total_sales = all_sales_data["order_value"].sum()
avg_order_value = all_sales_data["order_value"].mean()

if "product_category" in all_sales_data.columns:
    product_category_distribution = all_sales_data["product_category"].value_counts()
else:
    product_category_distribution = pd.Series(dtype=int)

print("Total Sales:", total_sales)
print("Average Order Value:", avg_order_value)
print("Product Category Distribution:\n", product_category_distribution)

# Step 8: Visualizations
if not product_category_distribution.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=product_category_distribution.index, y=product_category_distribution.values)
    plt.title("Sales by Product Category")
    plt.xlabel("Product Category")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

    plt.figure(figsize=(8, 8))
    plt.pie(product_category_distribution, labels=product_category_distribution.index, autopct="%1.1f%%")
    plt.title("Product Category Distribution")
    plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=all_sales_data, y="order_value")
plt.title("Order Value Distribution")
plt.ylabel("Order Value")
plt.show()
```
